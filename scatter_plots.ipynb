```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler 

data = {
    "age":[25,30,34,28,40],
    "height":[180,200,190,185,195],
    "salary":[20000,25000,20000,30000,28000]
}
df = pd.DataFrame(data)
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
print(scaled_df)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

data = {
    "age":[25,30,34,28,40],
    "height":[180,200,190,185,195],
    "salary":[20000,25000,20000,30000,28000]
}
df = pd.DataFrame(data)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
print(scaled_df)
```

```python
import matplotlib.pyplot as plt
# Sample data
temperature = [25, 28, 32, 35, 38, 40, 42, 45, 30, 33, 36, 37 ]# Temperature in Celsius
ice_cream_sales = [200, 250, 300, 350, 400, 420, 430, 450, 280, 310, 360, 370]# Sales in units
# Scatter plot=
plt.figure(figsize=(8, 6))
plt.scatter(temperature, ice_cream_sales, color='orange' , edgecolor='black' , alpha=0.8)
plt.title("Temperature vs. Ice Cream Sales", fontsize=14) 
plt.xlabel("Temperature (Â°C)", fontsize=12)
plt.ylabel("Ice Cream Sales (Units)", fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
data = {
'Temperature': [25, 28, 32, 35, 38, 40, 42, 45, 30, 33, 36, 37],
'Ice_Cream_Sales': [200, 250, 300, 350, 400, 420, 430, 450, 280, 310, 360, 370],
'Advertising_Budget': [50, 55, 60, 70, 75, 80, 85, 90, 60, 65, 70, 75], # In $100
'Customer_Count': [80, 100, 120, 140, 150, 160, 170, 180, 110, 130, 140, 145],
}
# Convert to DataFrame
df = pd. DataFrame(data)
# Pair plot
sns.pairplot(df, diag_kind='kde', markers="o", plot_kws={'alpha': 0.7})
plt.suptitle("Pair Plot of Ice Cream Sales Data", y=1.02, fontsize=16)
plt.show()
```
